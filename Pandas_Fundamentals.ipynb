## Missing Data Handling

Real-world datasets almost always have missing values. Pandas represents missing values as NaN (Not a Number). We already saw isnull() and isnull().sum() to detect them. Now we learn how to handle them properly.

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, np.nan, 22, 35, np.nan],
    "Salary": [50000, 70000, np.nan, 90000, 60000],
    "City": ["Delhi", "Mumbai", np.nan, "Pune", "Delhi"]
})

df
```

```python
# Check how many NaN values are in each column

df.isnull().sum()
```

```python
# dropna() → removes rows that have any NaN value

df.dropna() # Drops every row where at least one value is NaN
```

```python
# dropna() with how parameter

df.dropna(how='all') # Only drop a row if ALL values in that row are NaN (nothing dropped here since no row is all NaN)
```

```python
# dropna() along axis=1 drops columns that have any NaN

df.dropna(axis=1) # Drops every column that has at least one NaN
```

```python
# dropna() with thresh parameter

df.dropna(thresh=3) # Keep only rows that have at least 3 non-NaN values
```

```python
# fillna() → fills NaN values with a specific value

df.fillna(0) # Replace all NaN with 0
```

```python
# Fill each column with a different value

df.fillna({"Age": df["Age"].mean(), "Salary": df["Salary"].median(), "City": "Unknown"}) # Fill Age with mean, Salary with median, City with 'Unknown'
```

```python
# Forward fill (ffill) → fills NaN with the value from the previous row

df.ffill() # Each NaN gets the value from the row above it
```

```python
# Backward fill (bfill) → fills NaN with the value from the next row

df.bfill() # Each NaN gets the value from the row below it
```

```python
# interpolate() → fills NaN by interpolating between surrounding values

# This is very useful for time-series data like sensor readings or satellite data

df_ts = pd.DataFrame({
    "Displacement": [1.0, np.nan, 3.0, np.nan, 5.0] # Suppose these are InSAR displacement values
})

print(df_ts)
print(df_ts.interpolate()) # Linearly interpolates between 1.0 and 3.0 to fill the gap
```

## Data Type Conversion

We already saw how to set dtypes while reading a CSV. But often we need to convert column types after loading the data.

```python
df = pd.DataFrame({
    "ID": ["1", "2", "3", "4"],    # Stored as string
    "Score": ["85.5", "90.0", "abc", "78.3"], # Has a bad value 'abc'
    "Passed": [1, 0, 1, 1]           # Stored as int, but should be bool
})

print(df.dtypes)
df
```

```python
# astype() → convert column to a specific type

df["ID"] = df["ID"].astype(int) # Convert ID from string to int
df["Passed"] = df["Passed"].astype(bool) # Convert Passed from int to bool

print(df.dtypes)
df
```

```python
# pd.to_numeric() → convert column to numeric, with error handling

# astype() would raise an error on 'abc'. pd.to_numeric() with errors='coerce' converts bad values to NaN instead of crashing

df["Score"] = pd.to_numeric(df["Score"], errors='coerce') # 'abc' becomes NaN
print(df.dtypes)
df
```

## apply(), map() and transform()

These are the most important functions for applying custom operations to rows or columns. Instead of writing loops, you pass a function to these methods and Pandas applies it efficiently.

```python
df = pd.DataFrame({
    "Name": ["alice", "BOB", "Charlie", "david"],
    "Score": [85, 92, 78, 95],
    "Salary": [50000, 70000, 45000, 90000]
})

df
```

```python
# apply() on a column → applies a function to each element in the column

df["Name"] = df["Name"].apply(str.title) # Convert each name to title case
df
```

```python
# apply() with a custom function (lambda)

df["Grade"] = df["Score"].apply(lambda x: "Pass" if x >= 80 else "Fail") # Create a new column based on a condition
df
```

```python
# apply() on a full dataframe (axis=1 means apply row-wise)

# Suppose we want to create a new column that combines Name and Grade as a label
df["Label"] = df.apply(lambda row: row["Name"] + " - " + row["Grade"], axis=1) # axis=1 → process each row
df
```

```python
# apply() with axis=0 → apply function column-wise

print(df[["Score", "Salary"]].apply(lambda col: col.max() - col.min(), axis=0)) # Range (max - min) for each column
```

```python
# map() → applies a function or a dictionary mapping to a Series

# Useful for mapping categorical values to numbers (label encoding)

grade_map = {"Pass": 1, "Fail": 0} # Dictionary mapping
df["Grade_Numeric"] = df["Grade"].map(grade_map) # Map Grade column using the dictionary
df
```

```python
# transform() → applies a function but keeps the original shape (same index as original)

# Useful when you want to compute group-level stats but keep them aligned with the original dataframe

df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "Sales", "HR", "Sales"],
    "Salary": [70000, 50000, 80000, 60000, 55000, 65000]
})

# Add a column showing the average salary of each person's department (not just a summary — it stays aligned)
df["Dept_Avg_Salary"] = df.groupby("Department")["Salary"].transform("mean")
df
```

## Merging and Joining DataFrames

In real projects you rarely have all the data in one table. pd.merge() lets you combine two DataFrames based on a common column, similar to SQL joins. pd.concat() stacks DataFrames on top of each other or side by side.

```python
# Create two dataframes to merge

df_employees = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "DeptID": [10, 20, 10, 30]
})

df_departments = pd.DataFrame({
    "DeptID": [10, 20, 30, 40],
    "DeptName": ["IT", "HR", "Sales", "Finance"]
})

print(df_employees)
print(df_departments)
```

```python
# Inner Join → Only keeps rows that have matching DeptID in BOTH dataframes

pd.merge(df_employees, df_departments, on="DeptID", how="inner") # DeptID=40 disappears because no employee belongs to Finance
```

```python
# Left Join → Keeps all rows from the left dataframe, fills NaN where no match in right

pd.merge(df_employees, df_departments, on="DeptID", how="left") # All employees are kept
```

```python
# Right Join → Keeps all rows from the right dataframe, fills NaN where no match in left

pd.merge(df_employees, df_departments, on="DeptID", how="right") # Finance department appears with NaN for employee info
```

```python
# Outer Join → Keeps all rows from both dataframes, fills NaN where no match

pd.merge(df_employees, df_departments, on="DeptID", how="outer")
```

```python
# pd.concat() → stacks DataFrames vertically (on top of each other)

df1 = pd.DataFrame({"Name": ["Alice", "Bob"], "Score": [85, 90]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"], "Score": [78, 95]})

pd.concat([df1, df2]) # Stack df2 below df1
```

```python
# pd.concat() with ignore_index=True to reset the index

pd.concat([df1, df2], ignore_index=True) # Reset index to 0,1,2,3 instead of 0,1,0,1
```

```python
# pd.concat() horizontally (axis=1) → adds columns side by side

df_names = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie"]})
df_scores = pd.DataFrame({"Score": [85, 90, 78]})

pd.concat([df_names, df_scores], axis=1) # Place columns side by side
```

## GroupBy

```python
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "Sales", "HR", "Sales"],
    "Salary": [70000, 50000, 80000, 60000, 55000, 65000]
})

df
```

```python
# Suppose the question is what is the avg salary in each department?

df.groupby("Department")["Salary"].mean()  # Pandas internally creates groups -> Apply mean -> Combine
```

```python
# Count (Suppose we want how many employees in each department)

df.groupby("Department").count()
```

```python
# Sum of all employee salaries department wise

df.groupby("Department").sum()
```

```python
# Maximum salary in each department

df.groupby("Department").max()
```

## Aggregation

Aggregation functions reduce a column or group of values down to a single number. We already used agg() inside GroupBy. Here we look at aggregation on the full dataframe and custom aggregation functions.

```python
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "Sales", "HR", "Sales"],
    "Salary": [70000, 50000, 80000, 60000, 55000, 65000],
    "Age": [25, 30, 28, 35, 29, 40]
})

df
```

```python
# Applying aggregation on the whole dataframe (on numeric columns)

df[["Salary", "Age"]].agg(["mean", "min", "max", "sum", "count"])
```

```python
# Different aggregations for different columns using a dictionary

df.agg({"Salary": ["mean", "max"], "Age": ["min", "median"]})
```

```python
# Custom aggregation function with agg()

# Suppose we want the range (max - min) for each column
def range_func(x):
    return x.max() - x.min()

df[["Salary", "Age"]].agg(range_func)
```

```python
# Named aggregation with groupby → gives clean column names to aggregated results

df.groupby("Department").agg(
    Avg_Salary=("Salary", "mean"),  # Name the result column 'Avg_Salary'
    Max_Age=("Age", "max"),         # Name the result column 'Max_Age'
    Headcount=("Salary", "count")   # Name the result column 'Headcount'
)
```

## DateTime Handling

Time-series data is very common in geospatial work — sensor readings, satellite acquisition dates, climate data like ERA5. Pandas has excellent support for working with dates and times through the dt accessor and pd.to_datetime().

```python
df = pd.DataFrame({
    "Date": ["2024-01-15", "2024-03-22", "2024-06-10", "2024-09-05", "2024-12-31"],
    "Displacement": [1.2, 2.5, 3.1, 4.8, 5.3] # Suppose these are InSAR displacement values in mm
})

print(df.dtypes) # Date is stored as object (string) right now
df
```

```python
# pd.to_datetime() → converts a string column to datetime type

df["Date"] = pd.to_datetime(df["Date"])
print(df.dtypes) # Date is now datetime64
df
```

```python
# Once converted to datetime, we can use the dt accessor to extract date parts

print(df["Date"].dt.year)   # Extract year
print(df["Date"].dt.month)  # Extract month
print(df["Date"].dt.day)    # Extract day
print(df["Date"].dt.day_name()) # Get the day of the week
```

```python
# Set Date as the index (very useful for time-series)

df = df.set_index("Date")
df
```

```python
# Resample → aggregate data at a different time frequency

# Suppose we have daily data and want to compute monthly average displacement

df_daily = pd.DataFrame({
    "Displacement": [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5]
}, index=pd.date_range(start="2024-01-01", periods=12, freq="ME")) # ME = Month End frequency

print(df_daily)

# Resample to quarterly and compute mean
df_daily.resample("QE").mean() # QE = Quarter End
```

```python
# Rolling window → compute moving average over a sliding window

# A rolling mean smooths out noise in time-series data

df_daily["Rolling_Mean_3"] = df_daily["Displacement"].rolling(window=3).mean() # 3-period moving average
df_daily
```

## Window Functions

Window functions compute values over a sliding or expanding window. They are especially useful for time-series analysis — smoothing, trend detection, and cumulative statistics.

```python
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [100, 120, 90, 150, 130, 170]
})

df
```

```python
# rolling() → moving window

df["Rolling_Mean_3"] = df["Sales"].rolling(window=3).mean() # Average of the current and 2 previous values
df["Rolling_Max_3"] = df["Sales"].rolling(window=3).max()   # Max in each window of 3
df
```

```python
# expanding() → growing window (uses all data from the start up to the current row)

df["Expanding_Mean"] = df["Sales"].expanding().mean() # Average of all values up to and including the current row
df["Expanding_Sum"] = df["Sales"].expanding().sum()   # Cumulative sum up to current row
df
```

```python
# ewm() → Exponentially Weighted Mean (gives more weight to recent values)

# span controls how quickly the weights decay. Lower span = more weight on recent values.
df["EWM_Mean"] = df["Sales"].ewm(span=3).mean()
df
```

## Pivot Tables and Cross-tabulation

Pivot tables are a way to summarize data by rearranging it — think of them as a more flexible version of GroupBy that lets you spread values across both rows and columns. pd.crosstab() is a quick way to count occurrences of combinations of values.

```python
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "Sales", "HR", "Sales", "IT", "HR"],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Delhi", "Mumbai", "Pune", "Pune"],
    "Salary": [70000, 50000, 80000, 60000, 55000, 65000, 75000, 52000],
    "Gender": ["M", "F", "F", "M", "M", "F", "M", "F"]
})

df
```

```python
# Pivot table → average salary by Department (rows) and City (columns)

pd.pivot_table(df, values="Salary", index="Department", columns="City", aggfunc="mean")
```

```python
# Pivot table with multiple aggregation functions

pd.pivot_table(df, values="Salary", index="Department", aggfunc=["mean", "count", "sum"])
```

```python
# pd.crosstab() → counts how many times each combination of values appears

pd.crosstab(df["Department"], df["Gender"]) # How many Male and Female employees are in each department
```

```python
# crosstab with normalize=True gives proportions instead of counts

pd.crosstab(df["Department"], df["Gender"], normalize='index') # Shows proportion within each department (row)
```

- loc uses label-based indexing and iloc uses position-based indexing.
- Always check isnull().sum() first when loading a new dataset to understand missing data.
- Use dropna() to remove rows/columns with missing data, fillna() to fill them, and interpolate() for time-series gaps.
- sort_values() sorts by column values, sort_index() sorts by index labels.
- Adding a column is as simple as df['new_col'] = ..., and drop() removes columns (axis=1) or rows (axis=0).
- apply() and map() are the most powerful tools for custom transformations — avoid using for loops on DataFrames.
- transform() is like apply() but keeps the result aligned with the original DataFrame — essential for adding group-level features.
- pd.merge() is the SQL-style join for combining DataFrames, pd.concat() is for stacking them.
- GroupBy follows a split-apply-combine pattern and works best paired with agg().
- The str accessor handles all string operations vectorized, with automatic NaN handling.
- Always convert date columns with pd.to_datetime() before doing any time-based operations.
- rolling(), expanding(), and ewm() are essential for smoothing and trend analysis in time-series.
